```python
from sklearn.datasets import load_boston
from matplotlib import pyplot as plt
import numpy as np
boston=load_boston() 
```

```python
data = boston.data
target = boston.target
```

```python
NOX = data[:,4]
LSTAT = data[:,-1]
```

```python
#1
plt.scatter(LSTAT,NOX,color='#1E9E47')
plt.title("Exploring Boston Housing Data Relationships ")
plt.xlabel("Nitric oxides concentration(parts/10 million)")
plt.ylabel("%lower status of the population")
plt.show()
```

```python
#2

act_median = target * 1000
```

```python
z_axis = np.where(act_median > 20000,1,0)
```

```python
plt.scatter(LSTAT,NOX,c=z_axis,alpha=0.7)

cbar = plt.colorbar (ticks = [1,0])
cbar.set_ticklabels(['Median>$20,000','Median< $20,000'])

plt.title("Exploring Boston Housing Data Relationships ")
plt.xlabel("Nitric oxides concentration(parts/10 million)")
plt.ylabel("%lower status of the population")
plt.show()
```

```python
#3 
ZN = data[:,1]
ZN_1 = ZN[act_median > 20000]
ZN_2 = ZN[act_median <= 20000]
plt.hist(ZN_1, 10)
plt.hist(ZN_2, 10,alpha=0.8)

plt.title('Distribution of proportion of residential land')
plt.xlabel("Residential Land zoned for lots over 25,000 sq.ft.")
plt.ylabel("Number of observations")
plt.legend(['Median > $20,000','Median <= $20,000'])

plt.show()
```

```python
#4
LSAT= data[:,-1]
LSAT_1 = LSAT[act_median > 20000]
LSAT_2  = LSAT[act_median <= 20000]
plt.hist(LSAT_1, 10)
plt.hist(LSAT_2, 10,alpha=0.8)

plt.title('Distribution of Percentage of lower status of the population')
plt.xlabel("%lower status of the population")
plt.ylabel("Number of observations")
plt.legend(['Median > $20,000','Median <= $20,000'])

plt.show()
```
